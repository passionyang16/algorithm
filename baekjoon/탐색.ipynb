## 기본 탐색 알고리즘 - 기초 문제풀이

문제 제목: 문서검색

문제 난이도: 하

문제유형: 탐색

https://www.acmicpc.net/problem/1543

```python
#내 풀이 == 강사 풀이
word = input()
search = input()
count = 0
start = 0
while start <= len(word):
    if search == word[start:start+len(search)]:
        count += 1
        start += len(search)
    else:
        start += 1
    
print(count)
```

문제 제목: 베스트셀러

문제 난이도: 하

문제유형: 탐색

https://www.acmicpc.net/problem/1302

```python
#내 풀이
test_case = int(input())
dic = {}
for _ in range(test_case):
    data = input()
    if data in dic:
        dic[data] += 1
    else:
        dic[data] = 1

dic = dict(sorted(dic.items()))

print(max(dic.items(), key=lambda x:x[1])[0])
```

```python
#강사 풀이
n = int(input())
books = {}
for _ in range(n):
    book = input()
    if book not in books:
        books[book] = 1
    else:
        books[book] += 1

target = max(books.values())
array = []

for book, number in books.items():
    if number == target:
        array.append(book)

print(sorted(array)[0])
```

문제 제목: 트로피 진열

문제 난이도: 하

문제유형: 탐색

https://www.acmicpc.net/problem/1668

```python
#내 풀이
array = []
test_case = int(input())
for _ in range(test_case):
    array.append(int(input()))

left_highest = array[0]
left_count = 1
right_highest = array[-1]
right_count = 1

for i in range(1,len(array)):
    if array[i] > left_highest:
        left_highest = array[i]
        left_count += 1

for j in range(len(array)-2,-1,-1):
    if array[j] > right_highest:
        right_highest = array[j]
        right_count += 1

print(left_count)
print(right_count)
```

```python
#강사 풀이
def ascending(array):
    now = array[0]
    result = 1
    for i in range(1,len(array)):
        if now < array[i]:
            result += 1
            now = array[i]
    return result

n = int(input())
array = []
for _ in range(n):
    array.append(int(input()))

print(ascending(array))
array.reverse()
print(ascending(array))

```

## 기본 탐색 알고리즘 - 핵심 유형 문제풀이

문제 제목: 공유기 설치

문제 난이도: 중

문제유형: 이진 탐색

https://www.acmicpc.net/problem/2110

```python
# 내 풀이 - 개 똥
n,m = map(int,(input().split(' ')))
array = []
for _ in range(n):
    array.append(int(input()))

array = sorted(array)
result = []

def binary_search(array, m):
    if len(result) == m :
        return result
    
    result.append(array[0])
    m -= 1
    result.append(array[-1])
    m -= 1
        
    while True:
        medium = len(array) // 2
        result.append(array[medium])
        m -= 1
        
        if m == 0:
            break
        
        binary_search(array[:medium], m)
        binary_search(array[medium+1:], m)

binary_search(array, m)
```

```python
#강사 풀이
n, c = list(map(int, input().split(' ')))
array = []
for _ in range(n):
    array.append(int(input()))
array = sorted(array)

start = array[1] - array[0]
end = array[-1] - array[0]
result = 0

while (start <= end):
    mid = (start+end) // 2
    value = array[0]
    count = 1
    for i in range(1, len(array)):
        if array[i] >= value + mid:
            value = array[i]
            count += 1
    if count >= c:
        start = mid +1
        result = mid
    else:
        end = mid - 1

print(result)
            
```

문제 제목: 중량 제한

문제 난이도: 중상 

문제유형: 이진 탐색

https://www.acmicpc.net/problem/1939
    

```python
#강사 풀이.. 난 이런거 아직 못품 ㅠㅠ

from collections import deque

n, m = map(int,input().split())
adj = [[] for _ in range(n+1)]

# 중량을 찾는거임
def bfs(c):
    queue = deque([start_node])
    visited = [False] * (n+1)
    visited[start_node] = True
    while queue:
        x = queue.popleft()
        for y, weight in adj[x]:
            if not visited[y] and weight >= c:
                visited[y] = True
                queue.append(y)
    return visited[end_node]

start = 1000000000
end = 1
for _ in range(m):
    x, y, weight = map(int,input().split())
    #각 간선마다 추가해줌
    adj[x].append((y,weight))
    adj[y].append((x,weight))
    start = min(start,weight)
    end = max(end, weight)

start_node, end_node = map(int, input().split())

result = start

#여기는 이전 문제와 동일
while (start <= end):
    mid = (start+end) // 2
    if bfs(mid):
        result = mid
        start = mid +1
    else:
        end = mid - 1

print(result)
```

후위 순회: 왼쪽 자식 -> 오른쪽 자식 -> 루트

트리를 바르게 이해하기 위해선 재귀함수가 필요함

문제 난이도: 중

문제유형: 트리, 구현
    
https://www.acmicpc.net/problem/2250
    

```python
#강사 풀이 - 역대급 난이도인듯..하

class Node:
    def __init__(self, number, left_node, right_node):
        self.parent = -1
        self.number = number
        self.left_node = left_node
        self.right_node = right_node
        
def in_order(node, level):
    global level_depth, x
    level_depth = max(level_depth, level)
    if node.left_node != -1:
        in_order(tree[node.left_node], level + 1)
    level_min[level] = min(level_min[level], x)
    level_max[level] = max(level_max[level], x)
    x += 1
    if node.right_node != -1:
        in_order(tree[node.right_node], level + 1)

n = int(input())
tree = {}
level_min = [n]
level_max = [0]
root = -1
x = 1
level_depth = 1

for i in range(1, n+1):
    tree[i] = Node(i, -1, -1)
    level_min.append(n)
    level_max.append(0)
    
for _ in range(n):
    number, left_node, right_node = map(int, input().split())
    tree[number].left_node = left_node
    tree[number].right_node = right_node
    if left_node != -1:
        tree[left_node].parent = number
    if right_node != -1:
        tree[right_node].parent = number

for i in range(1, n+1):
    if tree[i].parent == -1:
        root = i

in_order(tree[root], 1)

result_level = 1
result_width = level_max[1] - level_min[1] + 1
for i in range(2, level_depth + 1):
    width = level_max[i] - level_min[i] + 1
    if result_width < width:
        result_level = i
        result_width = width

print(result_level, result_width)
```

```python
#내 풀이 - 이제부터 pypy3 사용하는걸로..
import heapq

array = []
result = [] 
n = int(input())
for _ in range(n):
    num = int(input())
    if not array and num == 0:
        result.append(0)
    elif num == 0:
        result.append(heapq.heappop(array))
    else:
        heapq.heappush(array, num)

for i in result:
    print(i)
```

```python
import heapq

n = int(input())
heap = []

for i in range(n):
    data = int(input())
    heapq.heappush(heap,data)

result = 0

while len(heap) != 1:
    one = heapq.heappop(heap)
    two = heapq.heappop(heap)
    sum_value = one + two
    result += sum_value
    heapq.heappush(heap, sum_value)

print(result)
```

문제 제목: 문제집

문제 난이도: 중

문제유형: 힙, 위상 정렬
    
https://www.acmicpc.net/problem/1766

위상 정렬 알고리즘

1) 진입차수가 0인 정점을 큐에 삽입합니다

2) 큐에서 원소를 꺼내 해당 원소와 간선을 제거합니다

3) 제거 이후에 진입차수가 0이 된 정점을 큐에 삽입합니다

4) 큐가 빌 때까지 2) - 3) 과정을 반복합니다

우선순위 큐라고도 불림

```python
import heapq

n, m = map(int, input().split())
array = [[] for i in range(n+1)]
indegree = [0] * (n+1)

heap = []
result = []

# 힙과 간선 구현
for _ in range(m):
    x, y = map(int, input().split())
    array[x].append(y)
    indegree[y] += 1

# 1번 0인 정점 큐에 삽입 구현
for i in range(1, n+1):
    if indegree[i] == 0:
        heapq.heappush(heap,i)

while heap:
    # 2번 큐에서 원소를 꺼냄
    data = heapq.heappop(heap)
    result.append(data)
    for y in array[data]:
        # 2번 간선을 제거
        indegree[y] -= 1
        # 3번 제거 이후에 진입차수가 0이 된 정점을 큐에 삽입
        if indegree[y] == 0:
            heapq.heappush(heap,y)

for i in result:
    print(i, end = ' ')
```
