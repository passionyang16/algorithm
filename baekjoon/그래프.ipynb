## 그래프 기본 탐색 알고리즘 - 기초 문제풀이

문제 제목: DFS와 BFS

문제 난이도: 하

문제유형: DFS, BFS
    
https://www.acmicpc.net/problem/1260

 정점 번호가 작은것 부터 방문
 
 큐 구현을 위해 deque를 사용

```python
#여러번 풀어봄...
from collections import deque

n, m, v = map(int, input().split())
adj = [[] for _ in range(n+1)]

for _ in range(m):
    x, y = map(int, input().split())
    adj[x].append(y)
    adj[y].append(x)

for e in adj:
    e.sort()

def dfs(v):
    print(v, end = ' ')
    visited[v] = True
    for e in adj[v]:
        if not (visited[e]):
            visited[e] = True
            dfs(e)

def bfs(v):
    q = deque([v])
    while q:
        v = q.popleft()
        if not(visited[v]):
            visited[v] = True
            print(v, end = ' ')
            for e in adj[v]:
                if not visited[e]:
                    q.append(e)

visited = [False] * (n+1)
dfs(v)
print()
visited = [False] * (n+1)
bfs(v)
```

문제 제목: 숨바꼭질

문제 난이도: 하

문제유형: BFS
    
https://www.acmicpc.net/problem/1697

```python
#내 풀이 : 왜 틀린거냐 진짜... 미치겠네
from collections import deque
n, k = map(int, input().split())

def bfs(n):
    visited = []
    height = 1
    q = deque([n])
    max_value = 2*n

    while n != k:
        n = q.popleft()
        if n == max_value:
            height += 1
            max_value = n
        if n not in visited:
            visited.append(n)
            q.extend([n-1,n+1,2*n])

    print(height)

bfs(n)
```

```python
#강사 풀이 - 선을 직접 구현
from collections import deque
MAX = 100001
n, k = map(int, input().split())
array = [0] * MAX

def bfs():
    q = deque([n])
    while q:
        now_pos = q.popleft()
        if now_pos == k:
            return array[now_pos]
        for next_pos in (now_pos-1, now_pos+1, now_pos * 2):
            #리스트 인덱싱으로 구현 if not array[next_pos]는 0일 때 진행한다는 의미
            if 0 <= next_pos < MAX and not array[next_pos]:
                # 이전에 연결된 거를 하나씩 더해서 저장해줌
                array[next_pos] = array[now_pos] +1
                q.append(next_pos)

print(bfs())
```

모든 정점에 대해 dfS, BFS 수행하고, 한번 방문한 정점은 다시 확인하지 않음

DFS 및 BFS를 수행한 총 횟수를 처리해야한다

DFS 경우 setrecursionlimit()함수를 설정해줘야한다

```python
import sys
sys.setrecursionlimit(100000)

def dfs(x,y):
    visited[x][y] = True
    directions = [(-1,0),(1,0),(0,-1),(0,1)]
    for dx, dy in directions:
        nx, ny = x+dx, y+dy
        if nx<0 or nx>=n or ny<0 or ny>=m:
            continue
        if array[nx][ny] and not visited[nx][ny]:
            dfs(nx,ny)

for _ in range(int(input())):
    m, n, k = map(int, input().split())
    array = [[0] * m for _ in range(n)]
    visited = [[False] * m for _ in range(n)]
    for _ in range(k):
        y, x = map(int, input().split())
        array[x][y] = 1
    result = 0
    for i in range(n):
        for j in range(m):
            if array[i][j] and not visited[i][j]:
                dfs(i,j)
                result += 1
    print(result)
```

모든 정점에 대해 DFS, BFS 수행

DFS 혹은 BFS 수행할때마다 방문하게 되는 노드의 개수를 계산

가장 노드의 개수가 크게 되는 시작 정점을 출력

```python
from collections import deque

n, m = map(int, input().split())
adj = [[] for _ in range(n+1)]

for _ in range(m):
    x, y = map(int, input().split())
    adj[y].append(x)

def bfs(v):
    q = deque([v])
    visited = [False] * (n+1)
    visited[v] = True
    count = 1
    while q:
        v = q.popleft()
        for e in adj[v]:
            if not visited[e]:
                q.append(e)
                visited[e] = True
                count +=1
    return count

result = []
max_value = -1

for i in range(1,n+1):
    c = bfs(i)
    if c > max_value:
        result = [i]
        max_value = c
    elif c == max_value:
        result.append(i)
        max_value = c

for e in result:
    print(e, end= ' ')
```

문제유형: 다익스트라 최단 경로
    
https://www.acmicpc.net/problem/10282

도달할 수 있는 정점들의 개수와 최대 거리

우선순위 큐를 이용하여 시간복잡도는 O(NlogD)로 해결 가능

```python
import heapq
import sys

def dijkstra(start):
    heap_data =[]
    #첫 노드를 힙에 저장
    heapq.heappush(heap_data, (0, start))
    # distance는 결국 지금까지 각 노드를 지날 때마다 최단 거리를 저장함
    distance[start] = 0
    while heap_data:
        #하나씩 꺼내서 간선 길이와 현재 노드를 반환
        dist, now = heapq.heappop(heap_data)
        # 간선길이가 최단거리보다 클 때: 걍 넘어감
        if distance[now] < dist:
            continue
        # 노드가 이어져있는 노드의 i[1]은 간선길이, i[0]는 이어진 노드
        for i in adj[now]:
            # 비용은 현재 거리에다가 간선을 더한 값
            cost = dist + i[1]
            # 그 더한 값으로 바꾼다음, 해당 최단 거리와 이어진 노드를 힙에 추가
            if distance[i[0]] > cost:
                distance[i[0]] = cost
                heapq.heappush(heap_data, (cost, i[0]))

for _ in range(int(input())):
    n, m, start = map(int, input().split())
    adj = [[] for i in range(n+1)]
    distance = [1e9] * (n+1)
    for _ in range(m):
        x, y, cost = map(int, input().split())
        adj[y].append((x, cost))
    dijkstra(start)
    count = 0
    max_distance = 0 
    for i in distance:
        if i != 1e9:
            #최단거리가 기록된 친구들만 반환하고 +1씩 해줌
            count += 1
            if i > max_distance:
                max_distance = i
    
    print(count, max_distance)
```

문제 제목: 우주신과의 교감

문제 난이도: 중

문제유형: 최소 신장 트리
    
https://www.acmicpc.net/problem/1774

2차원 좌표가 주어졌을 때, 모든 좌표를 잇는 최소 신장 트리 만들어야 함

2차원 좌표 상의 점을 잇는 통로들을 고려해야 함

크루스칼 알고리즘은 간선의 개수가 E 일 때 O(ElogE)로 동작

크루스칼 알고리즘으로 해결 할 수 있음

```python
import math

def get_distance(p1, p2):
    a = p1[0] - p2[0]
    b = p1[1] - p2[1]
    return math.sqrt((a*a) + b*b)

def get_parent(parent,n):
    if parent[n] == n:
        return n
    return get_parent(parent, parent[n])

def union_parent(parent, a, b):
    a = get_parent(parent, a)
    b = get_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b
    
def find_parent(parent, a, b):
    a = get_parent(parent, a)
    b = get_parent(parent, b)
    if a == b:
        return True
    else:
        return False

edges = []
parent = {}
locations = []
n, m = map(int, input().split())

for _ in range(n):
    x, y = map(int, input().split())
    locations.append((x,y))

length = len(locations)

#좌표마다 간선 설정. 이렇게 해야 0을 제외시키고 좌표끼리 길이 구할 수 있음
for i in range(length - 1):
    for j in range(i+1, length):
        edges.append((i+1, j+1, get_distance(locations[i],locations[j])))

#parent 하나씩 설정
for i in range(1,n+1):
    parent[i] = i

#하나 하나씩 parent를 찾아가는 과정
for i in range(m):
    a, b = map(int, input().split())
    union_parent(parent, a, b)

#최소 신장트리기 때문에 거리 기준으로 작은거 부터 정렬
edges.sort(key=lambda data: data[2])

result = 0
for a, b, cost in edges:
    #싸이클이 생기지 않을 때
    if not find_parent(parent, a, b):
        #parent 이어주고
        union_parent(parent, a, b)
        result += cost
        print(result)
            
print("%0.2f" % result)
```
